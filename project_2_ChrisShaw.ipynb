## Functions and Transformations

```python
def drop_columns (cols):
    ames_housing.drop(columns = cols, inplace=True)
    ames_housing_kaggle.drop(columns = cols, inplace=True)
```

```python
#values that are have less than a +- 0.1 correlation with saleprice
drop_columns(['pool_qc','misc_feature','alley','fence',
              'pid','3ssn_porch','mo_sold','pool_area',
              'bsmtfin_sf_2','misc_val','yr_sold','low_qual_fin_sf',
              'bsmt_half_bath','ms_subclass','overall_cond'])
```

```python
def fill_nulls ():
    ames_housing.fillna(0,inplace=True)
    ames_housing_kaggle.fillna(0,inplace=True)
```

## Making Dummy Columns the Same

```python
ames_housing_dummies = pd.get_dummies(ames_housing)
ames_housing_kaggle_dummies = pd.get_dummies(ames_housing_kaggle)
```

```python
missing_cols = set(ames_housing_dummies.columns ) - set(ames_housing_kaggle_dummies.columns)
    
for c in missing_cols:
    ames_housing_kaggle_dummies[c] = 0
        
ames_housing_kaggle_dummies = ames_housing_kaggle_dummies[ames_housing_dummies.columns]
```

```python
print(ames_housing_dummies.shape)
print(ames_housing_kaggle_dummies.shape)
```

## Heatmap Correlation Matrix

```python
plt.figure(figsize = (100,100))
ames_housing_dummies_corr = ames_housing_dummies.corr()
#ames_housing_dummies_corr[np.abs(ames_housing_dummies_corr)<0.6]=0
sns.heatmap(ames_housing_dummies_corr[['saleprice']].sort_values(by=['saleprice']),annot=True, vmin=-1, vmax=1, linewidths=.5, cmap='RdBu')
```

```python
plt.figure(figsize=(100,100))
sns.heatmap(ames_housing_dummies_corr[['saleprice']].sort_values(by=['saleprice']),mask=mask,annot=True, vmin=-1, vmax=1, linewidths=.5, cmap='RdBu')
```

```python
#making a list of low correlation columns to delete from dataframes

low_corr_cols = []

for row in range(len(mask)):
    if mask['saleprice'][row] ==True:
        low_corr_cols.append(mask.index[row])

if 'id' in low_corr_cols: low_corr_cols.remove('id')
```

```python
ames_housing_dummies.drop(columns = low_corr_cols, inplace = True)
ames_housing_kaggle_dummies.drop(columns = low_corr_cols, inplace = True)
```

## Feature Engineering

```python
#amplifying certain variables by putting them in the dataframe twice

ames_housing_dummies['1st_flr_sf_2'] = ames_housing_dummies['1st_flr_sf']
ames_housing_kaggle_dummies['1st_flr_sf_2'] = ames_housing_kaggle_dummies['1st_flr_sf']
```

```python
ames_housing_dummies['overall_qual_2'] = ames_housing_dummies['overall_qual']
ames_housing_kaggle_dummies['overall_qual_2'] = ames_housing_kaggle_dummies['overall_qual']
```

```python
ames_housing_dummies['gr_liv_area_2'] = ames_housing_dummies['gr_liv_area']
ames_housing_kaggle_dummies['gr_liv_area_2'] = ames_housing_kaggle_dummies['gr_liv_area']
```

```python
#multiplying overall quality by gr living area 
ames_housing_dummies['overall_qual_*gr_liv_area'] = ames_housing_dummies['overall_qual']*ames_housing_dummies['gr_liv_area']
ames_housing_kaggle_dummies['overall_qual_*gr_liv_area'] = ames_housing_kaggle_dummies['overall_qual']*ames_housing_kaggle_dummies['gr_liv_area']
```

```python
#multiplying overall quality by 1st flr sqft
ames_housing_dummies['overall_qual_*1st_flr_sf'] = ames_housing_dummies['overall_qual']*ames_housing_dummies['1st_flr_sf']
ames_housing_kaggle_dummies['overall_qual_*1st_flr_sf'] = ames_housing_kaggle_dummies['overall_qual']*ames_housing_kaggle_dummies['1st_flr_sf']
```

## Fitting Model

```python
from sklearn.model_selection import train_test_split,cross_val_score, KFold
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso,LassoCV
from sklearn.linear_model import ElasticNet,ElasticNetCV
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.metrics import r2_score

#all imports for modeling
```

```python
X = ames_housing_dummies.drop(columns='saleprice')
y = ames_housing_dummies['saleprice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=69, shuffle=True)
```

```python
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
#instantiate models
lr = LinearRegression()
lasso =LassoCV()
ridge = RidgeCV()
```

## Linear Regression

```python
lr.fit(X_train_sc,y_train);  
```

```python
lr.score(X_train_sc,y_train)
```

```python
lr.score(X_test_sc,y_test)
```

```python
cross_val_score(lr, X_train_sc,y_train).mean()
```

## Ridge Regression

```python
ridge.fit(X_train_sc,y_train);
```

```python
ridge.score(X_train_sc,y_train)
```

```python
ridge.score(X_test_sc,y_test)
```

```python
cross_val_score(ridge, X_train_sc,y_train).mean()
```

## Lasso Regression

```python
lasso.fit(X_train_sc,y_train);
```

```python
lasso.score(X_train_sc,y_train)
```

```python
lasso.score(X_test_sc,y_test)
```

```python
cross_val_score(lasso,X_train_sc,y_train).mean()
```
